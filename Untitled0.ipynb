```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("AmazonAnalysis") \
    .getOrCreate()
```

```python
df = spark.read.csv("/content/amazon.csv", header=True, inferSchema=True)
```

```python
df.show(5)
df.printSchema()
```

```python
from pyspark.sql.functions import col, desc, regexp_replace, when
```

```python
from pyspark.sql.functions import expr

# Clean actual price
df = df.withColumn(
    "actual_price_cleaned",
    expr("""
        try_cast(
            nullif(regexp_replace(actual_price, '[₹,]', ''), '')
        as float)
    """)
)

# Clean rating count
df = df.withColumn(
    "rating_count_cleaned",
    expr("""
        try_cast(
            nullif(regexp_replace(rating_count, ',', ''), '')
        as int)
    """)
)

# Clean discount %
df = df.withColumn(
    "discount_percentage_cleaned",
    expr("""
        try_cast(
            nullif(regexp_replace(discount_percentage, '[^0-9]', ''), '')
        as int)
    """)
)

# Clean rating
df = df.withColumn(
    "rating_cleaned",
    expr("""
        try_cast(nullif(rating, '') as float)
    """)
)
```

```python

# =====================================================
# 1. Top 10 Most Expensive Products
# =====================================================

df.orderBy(desc("actual_price_cleaned")) \
  .select("product_name", "actual_price_cleaned") \
  .show(10)
```

```python
# =====================================================
# 2. Top Selling Category (based on highest reviews)
# =====================================================

df.groupBy("category") \
  .sum("rating_count_cleaned") \
  .withColumnRenamed("sum(rating_count_cleaned)", "total_reviews") \
  .orderBy(desc("total_reviews")) \
  .show(10)
```

```python
# =====================================================
# 3. Price Distribution Analysis
# =====================================================

df.select("actual_price_cleaned").describe().show()
```

```python
df.orderBy(desc("actual_price_cleaned")).show(10)
```

```python
# ==========================================================
# STEP 4: PRICE DISTRIBUTION ANALYSIS
# ==========================================================

# Get statistical summary of prices
df.select("actual_price_cleaned").describe().show()
```

```python
# ==========================================================
# STEP 5: PRODUCTS WITH HIGHEST DISCOUNT
# ==========================================================

# Sort by highest discount percentage
df.orderBy(desc("discount_percentage_cleaned")) \
  .select("product_name", "discount_percentage_cleaned") \
  .show(10, False)
```

```python
# ==========================================================
# STEP 6: CORRELATION (PRICE vs RATING)
# ==========================================================

# Find correlation between price and rating
correlation = df.select("actual_price_cleaned", "rating_cleaned") \
                .dropna() \
                .stat.corr("actual_price_cleaned", "rating_cleaned")

print("Correlation between Price and Rating:", correlation)
```
