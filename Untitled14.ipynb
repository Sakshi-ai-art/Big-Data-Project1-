```python
import pandas as pd

df = pd.read_csv("amazon.csv")
df.head()
```

```python
#NUMBER OF COLUMNS IN DATASET
df.columns
```

```python
#SELECT COLUMNS
df[["product_name", "category", "discounted_price"]].head(10)
```

```python
#FILTER DATA
df["discounted_price"]
```

```python
#COUNT ROWS
len(df)
```

```python
#CLEANING PRICE COLUMNS
df["discounted_price"] = df["discounted_price"] \
    .replace('[₹,]', '', regex=True) \
    .astype(float)

df["actual_price"] = df["actual_price"] \
    .replace('[₹,]', '', regex=True) \
    .astype(float)
```

```python
#TRANSFORMATION
df["final_price"] = df["discounted_price"] * 0.9
df.head()
```

```python
#GROUP BY AGGREGATION
df.groupby("category")["product_name"].count()
```

```python
#SORTING
df.sort_values(by="discounted_price", ascending=False).head(10)
```

```python
#MAP PHASE
mapped = [(row["category"], 1) for index, row in df.iterrows()]
```

```python
#REDUCE PHASE
from collections import defaultdict

result = defaultdict(int)

for key, value in mapped:
    result[key] += value

result
```

```python
#AVERAGE PRICE
mapped = [(row["category"], (row["discounted_price"], 1)) for index, row in df.iterrows()]

result = defaultdict(lambda: [0, 0])

for key, (price, count) in mapped:
    result[key][0] += price
    result[key][1] += count

average_price = {k: v[0]/v[1] for k, v in result.items()}

average_price
```

```python
#ADVANCED AGGREGATION
df["rating"] = pd.to_numeric(df["rating"], errors='coerce')

df["rating_count"] = df["rating_count"] \
    .replace(',', '', regex=True)

df["rating_count"] = pd.to_numeric(df["rating_count"], errors='coerce')
```

```python
result = df.groupby("category").agg({
    "rating": "mean",
    "rating_count": "sum"
})

result = result.rename(columns={
    "rating": "avg_rating",
    "rating_count": "total_reviews"
})

result.head(10)
```
